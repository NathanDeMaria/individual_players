Use the vpp model to set a prior for all players and update them based on all performances

```python
%load_ext autoreload
%autoreload 2
```

```python
from individual_players import LeagueModel, build_combined_df, update_loop
```

```python
gender = "womens"
df = build_combined_df(gender, verbose=False)
model = LeagueModel.load(f"models/{gender}_league.pkl")
df = df.assign(vpp_sd=model.possessions_to_vpp_std(df.n_possessions))

# TODO: actually sort by date
df = df.sort_values('game_id')
```

```python
player_ratings_df = update_loop(df, model)
```

```python
# Check out the best players
with_ratings = (
    df.groupby('player_id').agg({"n_possessions": "sum"})
    .reset_index()
    .merge(player_ratings_df, left_on='player_id', right_index=True)
)
for row in with_ratings[with_ratings.n_possessions > 2000].sort_values('vpp', ascending=False).itertuples():
    print(f"https://www.espn.com/{gender}-college-basketball/player/_/id/{row.player_id}", row.vpp)
```
